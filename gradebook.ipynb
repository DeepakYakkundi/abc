### hw_exam_df Dataframe

```python
hw_exam_df = (
    pd.read_csv("data-sets/gradebook_data/hw_exam_grades.csv")
    .rename(columns = lambda label : str(label).lower().replace(" ", "_").replace("_-_","_"))
    .assign(homework_1 = lambda df : df["homework_1"].fillna(0).astype(dtype = {"homework_1" : np.int64}))
    )

# req_columns = [col for col in hw_exam_df.columns if "_time" not in col]
# hw_exam_df = hw_exam_df[req_columns]

hw_exam_df = hw_exam_df.filter(regex = r'\b(?!.*_time).+\b', axis = 1)
```

```python
hw_exam_df.info()
```

### quiz_df Dataframe

```python
from pathlib import Path
from functools import reduce

# Create a Path object for your directory
base_path = Path("./data-sets/gradebook_data/")

# Use .glob() for non-recursive matching in the current directory
# Pattern to find files that start with "report"
files_matching_name = list(base_path.glob("quiz_*"))
# print(files_matching_name)

# Use .rglob() for recursive matching (all subdirectories)
# Pattern to find all files ending with ".csv" in the directory tree
recursive_files = list(base_path.rglob("quiz_*"))
# print(recursive_files)

dfs = []

# Print the list of Path objects (which can be used as strings)
for idx, file_path in enumerate(files_matching_name):
    # print(file_path)
    df_name = f"quiz{idx+1}_df"
    # dfs.append(df_name)
    grade_col = f"quiz_grade_{idx+1}"
    df_name = pd.read_csv(file_path, header = None, skiprows=1,
                       names = ["last_name","first_name","email", grade_col],
                       dtype = {
                           grade_col : np.int8
                           })
    dfs.append(df_name)


# Merge all DataFrames on the 'name' column using an outer join
quiz_df = reduce(lambda left, right: pd.merge(left, right, on=['last_name', 'first_name', 'email' ], how='outer'), dfs)

print(quiz_df)
```

### Final Dataframe. All operations will be done on this

```python
df = pd.merge(hw_exam_df, quiz_df, on=['last_name', 'first_name'], how='outer')
df.info()
```

### calculations

```python
# Calculate homework average

df["total_hw_score"] = np.sum(df.filter(regex = r'\bhomework_\d?\b', axis = 1), axis = 1)
df["max_hw_score"] = np.sum(df.filter(regex = r'homework_\d_max_points\b', axis = 1), axis = 1)
df["hw_avg"] = df["total_hw_score"] / df["max_hw_score"]
```

```python
# Calculate Exam average

df["total_exam_score"] = np.sum(df.filter(regex = r'\bexam_\d?\b', axis = 1), axis = 1)
df["max_exam_score"] = np.sum(df.filter(regex = r'exam_\d_max_points\b', axis = 1), axis = 1)
df["exam_avg"] = df["total_exam_score"] / df["max_exam_score"]

```

```python
# Calculate Quiz average

df["total_quiz_score"] = np.sum(df.filter(regex = r'\bquiz_grade_\d?\b', axis = 1), axis = 1)
df["max_quiz_score"] = df.filter(regex = r'\bquiz_grade_\d?\b', axis = 1).max().sum()
df["quiz_avg"] = df["total_quiz_score"] / df["max_quiz_score"]
```

```python
weights = pd.Series(
    index = ["hw_avg", "quiz_avg", "exam_avg"],
    data = [30, 20, 50]
)
```

```python
df["final_score"] = np.sum(df[weights.index] * weights, axis = 1)
```

```python
df["final_score"].describe()
```

```python
def calculate_grade(ser):
    if ser > 80:
        return "A"
    elif ser > 70:
        return "B"
    else:
        return "C"
```

```python
df["final_grade"] = df["final_score"].apply(calculate_grade)
```

```python
df["final_grade"].value_counts()
```

```python
df[df["final_grade"] == "A"].to_csv("student_grades_A.csv")
df[df["final_grade"] == "B"].to_csv("student_grades_B.csv")
df[df["final_grade"] == "C"].to_csv("student_grades_C.csv")
```
