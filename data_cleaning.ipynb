```python
with open("data-sets/university_towns.txt", "r") as file:
    state = ""
    university_rows = []
    for line in file:
        if line.strip().endswith("[edit]"):
            state = str(line).split("[")[0]
            # print(state)
        elif line.strip().endswith(":"):
            state = str(line).split(":")[0]
        else:
            town = line.split(" (")[0]
            universities = line.split(" (")[1]
            # print(f"  town: {town}")
            # print(f"      universities: {universities}")

        if town and universities:
            university_rows.append([state, town, universities])
            town = ""
            universities = ""

# print(university_rows)
        
universities_df = pd.DataFrame(data = university_rows,
             columns=["state", "town", "universities"])
```

```python
def read_file():
        
    with open("data-sets/university_towns.txt", "r") as file:
        state = ""
        university_rows = []
        for line in file:
            if "[edit]" in line:
                state = line.strip()
            else:
                town = line.strip()
                university_rows.append([state, town])

    return pd.DataFrame(data = university_rows, columns = ["state", "town"])
```

```python
university_df = read_file()
university_df.to_csv("towns.csv")
```

```python
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
display(df)
# The index and columns initially have no name
print("Original DataFrame index name:", df.index.name)
print("Original DataFrame columns name:", df.columns.name)

# Rename the index axis to 'Row Labels'
df_renamed_index = df.rename_axis(index='Row Labels')
display(df_renamed_index)
# Rename the columns axis to 'Column Features'
df_renamed_columns = df.rename_axis(columns='Column Features', index='Row Labels')
display(df_renamed_columns)
print("\nDataFrame with renamed index axis name:")
print(df_renamed_index.index.name) # Output: Row Labels

print("\nDataFrame with renamed columns axis name:")
print(df_renamed_columns.columns.name) # Output: Column Features
```

```python
def fruit_cat(s):
    # print(s)
    # return 'sweet'
    if s.iloc[1] == 'Apple':
        return 'Sweet'
    elif s.iloc[1] == 'Pear':
        return 'Sour'
```

```python
def categorize_temp(row):
    if row['temp_c'] <= 17:
        return 'cool'
    elif row['temp_c'] <= 25:
        return 'warm'
    else:
        return 'hot'
```
